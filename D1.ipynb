```python
import numpy as np
import matplotlib.pyplot as plt
# Day 1 â€” Linear Regression (From Scratch)
#oal: Implement gradient descent to fit a line to synthetic data.
# y = 3x + 2 + noise

np.random.seed(0)
X = np.linspace(0, 10, 100)
noise = np.random.randn(100)
y = 3 * X + 2 + noise

plt.scatter(X, y)
plt.title("Synthetic Data")
plt.show()
```

```python
m = 0.0  # slope
b = 0.0  # intercept
lr = 0.001  # learning rate
```

```python
for epoch in range(500):
    y_pred = m * X + b
    error = y_pred - y

    dm = (2/len(X)) * np.dot(error, X)
    db = (2/len(X)) * np.sum(error)

    m -= lr * dm
    b -= lr * db

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: m={m:.3f}, b={b:.3f}")
```

```python
plt.scatter(X, y, label="data")
plt.plot(X, m * X + b, color="red", label="fitted line")
plt.legend()
plt.title("Final Fitted Line")
plt.show()
```
